```python
import pandas as pd
df = pd.read_csv ('loan_data.csv')
print(df.dtypes)  
```

```python
df.head(5)
```

```python
loan_data = df[['purpose','installment', 'int.rate', 'fico', 'revol.bal','not.fully.paid' ]]
loan_data.head(5)
```

```python
# Lets check the count of missing values in the dataframe
loan_data.isna().sum()
```

```python
# Lets see the heatmap for the correlation among the attributes.
import seaborn as sns
corr = loan_data.corr()
# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
hist = loan_data.hist(bins=20, figsize=(15, 10))
```

```python
loan_data.describe(include='all')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
sns.countplot(x='purpose',data=loan_data, palette='rainbow',hue='not.fully.paid')
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=loan_data, x="int.rate", y="fico")
```

```python
#Hypothisis testing on fico score and interest rate are dependent on each other.

#H0 : fico score and interest rate are dependent on each other.
#H1 : fico score and interest rate are not dependent on each other.

from scipy.stats import pearsonr
stat, p = pearsonr(loan_data['int.rate'], loan_data['fico'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
```
